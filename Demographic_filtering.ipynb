```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv('dataset/tmdb_5000_credits.csv')
data2=pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
data1.head()
```

```python
data2.head()
```

```python
columns=['id','title','cast','crew']
data1.columns=columns
data2=data2.merge(data1,on='id')
```

```python
data2.head()
```

```python
c=data2['vote_average'].mean()
```

```python
c
```

```python
m=data2['vote_count'].quantile(0.9)
m
```

```python
copy=data2.copy()
q_movies=copy[copy['vote_count']>=m]
```

```python
q_movies.shape
```

```python
def weighted_rating(data2,m=m,C=c):
    v=data2['vote_count']
    R=data2['vote_average']
    return(v/(v+m)*R)+(m/(m+c)*c)
```

```python
q_movies['score']=q_movies.apply(weighted_rating,axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)
q_movies[['title_x', 'vote_count', 'vote_average', 'score']].head(10)
```
